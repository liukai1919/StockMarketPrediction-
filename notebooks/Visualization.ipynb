```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/processed_data_news_stock.csv')
df.head()
```

```python
# show top 10 change
df[df['company'] == 'AAPL'].sort_values(by='change', ascending=False).head(50)
```

```python
# plot the relationship between change and label
# Set the style of the plot
sns.set_style("whitegrid")

# Create the scatter plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    x='change', 
    y='label', 
    data=df[df['pubDate'].notna()],
    hue='label',  # Color by label if it's categorical
    palette='viridis',  # Use a color palette
    s=100,  # Marker size
    alpha=0.7  # Transparency
)

# Add titles and labels
plt.title('Scatter Plot of Change vs Label')
plt.xlabel('Change')
plt.ylabel('Label')

# Add a legend
plt.legend(title='Label', loc='upper right')

# Show the plot
plt.show()
```

```python
#plot the relationship between change and score
# Create the scatter plot
plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(
    x='score', 
    y='change', 
    data=df[df['pubDate'].notna()],
    hue='score',  # Color by score if it's categorical or continuous
    palette='coolwarm',  # Use a color palette
    s=80,  # Marker size
    alpha=0.6,  # Transparency
    edgecolor='w',  # White edge color for better separation
    linewidth=0.5  # Edge line width
)

# Add titles and labels
plt.title('Scatter Plot of Change vs Score', fontsize=16)
plt.xlabel('Score', fontsize=14)
plt.ylabel('Change', fontsize=14)

# Add a legend
plt.legend(title='Score', loc='upper right', fontsize=12)

# Optionally add a trend line
sns.regplot(
    x='change', 
    y='score', 
    data=df[df['pubDate'].notna()],
    scatter=False,  # Don't plot the scatter again
    color='gray',  # Trend line color
    line_kws={'linewidth': 1.5, 'linestyle': '--'}  # Line style
)

# Show the plot
plt.show()
```
